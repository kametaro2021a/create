```python
f = open("newfile.txt", "w", encoding="utf-8")
f.write("人生はゼロから生まれゼロに戻るのだから、失敗しても何も損はしない")
f.close()
```

`newfile.txt`というテキストファイルを`w`(writeのwだから書き込む)というオプションで，尚且つ文字エンコードは`utf-8`で開いて(`open`)それを`f`としておく。

そして，その`f`とした「`utf-8`で書き込むために開いた`newfile.txt`」に書き込み(`f.write("<書き込みたい文字列>")`)，書き込んだからファイルをちゃんと「閉じて」(`f.close()`)おく。

そういうプログラムです。ではこのファイルはどこに保存されたのでしょう。

この画面（colaboratoryが動いてるブラウザの画面）の左上にオレンジのColaboratoryのアイコンがありその下に4つのグレーのアイコンが並んでる4番目が「フォルダ」のマークになっています。そこを押してみると判ります。

押すと，この画面の左にファイルブラウザの細長い表示が追加され，Colaboratoryの編集画面は少し狭く表示されます。

もちろんもう一度押せば「ファイルブラウザ」部分は消えて元通りになります。何度か試してみて納得しておきましょう。


先ずはこの`newfile.txt`が置かれたところに何かテキストファイルを置いて読み込んでみます。今回は青空文庫から落とした「おしゃべり姫」というお話をちゃんと解凍したテキストファイルをドラッグ＆ドロップで置いてみました。

```python
# テキストファイルを開いて一行ずつ読み込んで表示する。
f = open("newfile.txt", "r", encoding="utf-8")
for line in f:             # ファイルから一行ずつ読み込む
    print(line, end=" ")   # 一行ずつ表示する
f.close()
```

青空文庫のテキストファイルは`shift-jis`というエンコードで作成されているので，オプション部分のエンコードを変えて開いてみる。

```python
# テキストファイルを開いて一行ずつ読み込んで表示する。
f = open("oshaberihime.txt", "r", encoding="shift_jis")
for line in f:             # ファイルから一行ずつ読み込む
    print(line, end=" ")   # 一行ずつ表示する
f.close()
```

次のような開き方もある。あと`.rstrip()`を施すことで，テキストファイルの一行毎の右端に付いたもの（改行コードとか？）が削除されるようだ。

```python
# テキストファイルを開いて読み込んで表示する。
with open('oshaberihime.txt', encoding='shift_jis') as f:
    s = f.read()
#print(s)
print(s.rstrip().rstrip())
```

そんなわけで，`shift-jis`なテキストファイルを読み込んで`utf-8`で新しくファイルに書き込むという事を試してみた。

```python
f = open('oshaberihime.txt', 'r', encoding='shift_jis')
g = open('oshaberihimeA.txt', 'w', encoding='utf-8')
for i in range(1,17):
  line = f.readline()
for i in range(1,10):
  line = f.readline()
#  print(line.strip())
  g.write(line.rstrip()+"\n")
f.close()
g.close()
f = open('oshaberihimeA.txt', 'r', encoding='utf-8')
for i in range(1,10):
  line = f.readline()
  print(line.rstrip())
f.close()
```

```python
import re # 正規表現ライブラリ
mecab = MeCab.Tagger("-Owakati")
f = open('oshaberihime.txt', 'r', encoding='shift_jis')
g = open('oshaberihimeA.txt', 'w', encoding='utf-8')
for i in range(1,17):
  line = f.readline()
for i in range(1,503):
  line = f.readline().rstrip()
  line=re.sub(r'《.*?》|［＃.*?］|｜', '', line)
  line=line.strip('「')	
  line=line.strip('」')	
  line=mecab.parse(line).lstrip()
#  oshaberi = markovify.Text(line.rstrip())
#  print(line.strip())
  g.write(line.rstrip()+"\n")
f.close()
g.close()
f = open('oshaberihimeA.txt', 'r', encoding='utf-8')
line = f.readline()
while line:
#  print(line.rstrip())
  line = f.readline()
f.close()
```
